```python
part = sBach.parts[0]

for element in part:
    if type(element) is stream.Measure:
        for el in element:
            print '---start item---'
            print el
            if type(el) is note.Note:
                print 'Pitch Class'
                print el.pitch.pitchClass
                print 'Octave'
                print el.pitch.octave
                print 'Length Relative to Quarter'
                print el.duration.quarterLength
                if el.tie is not None:
                    print el.tie
            elif type(el) is note.Rest:
                print 'Length Relative to Quarter'
                print el.duration.quarterLength
            print '---end item---'
    else:
        print '---start item---'
        print element
        print '---end item---'
```
